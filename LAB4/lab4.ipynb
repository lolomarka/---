```python
# Напишем функцию для преобразования столбцов с "нестандартными флагами" к булевому виду

def custom_bool_to_bolean_column(data_column, true_str, false_str):
    if len(data_column.unique()) > 2:
        raise ValueError('data_column must contain only two unique values') 
    return data_column.map({true_str : True, false_str: False}).astype(bool)

telecom_churn['International plan'] = custom_bool_to_bolean_column(telecom_churn['International plan'], 'Yes', 'No')
telecom_churn['Voice mail plan'] = custom_bool_to_bolean_column(telecom_churn['Voice mail plan'], 'Yes', 'No')
```

```python
# 8. Добавьте в датафрейм столбец «Прогнозируемый отток», заполнив его на основе
# значений столбцов «Customer service calls» и «International plan». Сравните
# значение в этом столбце со значением столбца «Churn». Если мы будем
# пользоваться построенным прогнозом, то какой процент ошибок первого и
# второго рода (ложноположительных и ложноотрицательных) мы получим?

def predict_churn(row):
    if row['Customer service calls'] >= 4 or row['International plan'] == True:
        return True
    else:
        return False
    
telecom_churn['Predicted Churn'] = telecom_churn.apply(predict_churn, axis=1)
telecom_churn.head(10)
```

```python
# Сравнение прогнозируемого оттока с настоящим
# Ложноположительные ошибки (False Positives)
false_positives = ((telecom_churn['Predicted Churn'] == True) & (telecom_churn['Churn'] == False)).sum()

# Ложноотрицательные ошибки (False Negatives)
false_negatives = ((telecom_churn['Predicted Churn'] == False) & (telecom_churn['Churn'] == True)).sum()

# Общее количество клиентов
total_clients = len(telecom_churn)

# Процент ложноположительных и ложноотрицательных ошибок
false_positive_rate = false_positives / total_clients * 100
false_negative_rate = false_negatives / total_clients * 100

# Вывод результатов
print(f"Ложноположительные ошибки (False Positives): {false_positives} ({false_positive_rate:.2f}%)")
print(f"Ложноотрицательные ошибки (False Negatives): {false_negatives} ({false_negative_rate:.2f}%)")
```

```python
# 5) Рассмотрим гонщиков, которые становились чемпионами хотя бы дважды.
# Рассмотрим промежутки времени между их «соседними» чемпионствами.
# Каков максимальный перерыв между двумя последовательными
# чемпионствами одного и того же гонщика? Например, Нельсон Пике
# становился чемпионом в 1981, 1983 и 1987 годах. Для него такие перерывы
# составляют 2 года (с 1981 до 1983) и 4 года (с 1983 до 1987), максимальный
# из них – 4 года. Вам нужно найти максимум среди всех гонщиков. (В этом
# задании можно не ограничиваться только функциями библиотек, но и
# написать свой код на Python.)

at_least_two_wins_racers = champions_df[champions_df.Wins['Wins'] >= 2]
at_least_two_wins_racers
```

```python
at_least_two_wins_racers.columns = at_least_two_wins_racers.columns.droplevel()
```

```python
champions_years = at_least_two_wins_racers.groupby('Driver').agg(
    { 'Season' : list}
)
champions_years
```

```python
def find_max_gap(row):
    max_gap = 0
    years = row.Season
    years.sort()
    gaps = [years[i] - years[i-1] for i in range(1, len(years))]
    if gaps:
        max_gap = max(max_gap, max(gaps))

    return max_gap
```

```python
champions_years['Max gaps'] = champions_years.apply(find_max_gap, axis=1)
```
