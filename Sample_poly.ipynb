```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('data2.csv', sep = ';')
df.head()
```

```python
plt.figure(figsize = (6,4))
plt.scatter(df.proc, df.income)
```

```python
X = ['proc']
Y = 'income'
X_train, X_test, Y_train, Y_test = train_test_split(df[X], df[Y], test_size = 0.3, random_state = 142)
```

```python
plt.scatter(X_train, Y_train)
plt.scatter(X_test, Y_test, color = 'red')
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, Y_pred)
plt.scatter(X_test, Y_test, color = 'red')
```

```python
print('RMSE:', root_mean_squared_error(Y_test, Y_pred))
print('R2:', np.round(r2_score(Y_test, Y_pred), 2))
```

```python
pf=PolynomialFeatures(degree = 2)     
X_train_poly=pf.fit_transform(X_train)
X_test_poly=pf.fit_transform(X_test)
model2 = linear_model.LinearRegression()
model2.fit(X_train_poly, Y_train)
Y_pred2 = model2.predict(X_test_poly)
```

```python
plt.scatter(X_test, Y_pred2)
plt.scatter(X_test, Y_test, color = 'red')
```

```python
print('RMSE:', root_mean_squared_error(Y_test, Y_pred2))
print('R2:', np.round(r2_score(Y_test, Y_pred2), 2))
```

```python
pf3=PolynomialFeatures(degree = 3)     
X_train_poly3=pf3.fit_transform(X_train)
X_test_poly3=pf3.fit_transform(X_test)
model3 = linear_model.LinearRegression()
model3.fit(X_train_poly3, Y_train)
Y_pred3 = model3.predict(X_test_poly3)
```

```python
plt.scatter(X_test, Y_pred3)
plt.scatter(X_test, Y_test, c='red')
```

```python
print('RMSE:', root_mean_squared_error(Y_test, Y_pred3))
print('R2:', np.round(r2_score(Y_test, Y_pred3), 5))
```
